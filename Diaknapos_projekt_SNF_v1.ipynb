# Hány Diáknapon vettek részt a 2006-ban induló csapatok, valamint az elért pontszámuk:

# Csapatok részvételének számításai és jóslatai

```python
import pandas as pd

dfs = []
for i, df in enumerate([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15, df16]):
    df = df.copy()
    df['ev'] = 2006 + i
    dfs.append(df)

df_osszes = pd.concat(dfs, ignore_index=True)
```

```python
import unicodedata
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'DejaVu Sans'

def remove_problematic_chars(text):
    return ''.join(c for c in text if ord(c) != 128 and unicodedata.category(c)[0] != 'C')

reszvetelek = df_osszes.groupby('csapat')['ev'].nunique().reset_index()
reszvetelek.columns = ['csapat', 'reszvetelek_szama']
reszvetelek['csapat'] = reszvetelek['csapat'].apply(remove_problematic_chars)
reszvetelek = reszvetelek.sort_values(by='reszvetelek_szama', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=reszvetelek, x='csapat', y='reszvetelek_szama', hue='csapat', palette='viridis')

plt.title('Csapatok részvételeinek száma az összes év során', fontsize=14)
plt.xlabel('Csapat', fontsize=12)
plt.ylabel('Részvételek száma (évek)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Az ábra azt mutatja meg, hogy az egyes csapatok hány különböző évben vettek részt a vizsgált időszak során (2006-tól kezdődően). A függőleges oszlopok magassága azt jelzi, hogy az adott csapat hány évben volt jelen a diáknapi eseményeken. Ezáltal jól látható, hogy mely csapatok voltak a legaktívabbak, vagyis melyek képviseltették magukat a legtöbb évben.

```python
from sklearn.linear_model import LinearRegression

kezdo_ev = 2006
df_list = []

# Összefűzés
for i in range(1, 17):
    ev = kezdo_ev + i - 1
    df = globals()[f"df{i}"].copy()
    df['ev'] = ev
    df_list.append(df)

df_all = pd.concat(df_list, ignore_index=True)

# Lineáris regresszió
csapatok = df_all['csapat'].unique()
predictions = []

for csapat in csapatok:
    team_data = df_all[df_all['csapat'] == csapat].copy()
    team_data = team_data.dropna(subset=['pont', 'ev'])

    if len(team_data) >= 3:
        X = team_data[['ev']]
        y = team_data['pont']

        if y.isnull().any() or X.isnull().values.any():
            continue

        model = LinearRegression()
        model.fit(X, y)
        pred_2026 = model.predict(pd.DataFrame({'ev': [2026]}))[0]

        predictions.append({
            'csapat': csapat,
            'predikcio_2026': round(pred_2026, 2),
            'reszvetelek_szama': len(team_data),
            'osszpont': team_data['pont'].sum()
        })

df_pred = pd.DataFrame(predictions)
df_pred_sorted = df_pred.sort_values(by='predikcio_2026', ascending=False)
top5 = df_pred_sorted.head(5)

# Ábra
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=top5, x='csapat', y='predikcio_2026', hue='csapat', palette='crest', legend=False)

plt.title('Top 5 csapat várható pontszáma 2026-ban (Lineáris regresszió alapján)', fontsize=14)
plt.xlabel('Csapat', fontsize=12)
plt.ylabel('Várható pontszám', fontsize=12)
plt.ylim(0, top5['predikcio_2026'].max() + 10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)

# Pontszám kiírás
for i, row in enumerate(top5.itertuples()):
    ax.text(x=i, y=row.predikcio_2026 + 0.5, s=f"{row.predikcio_2026}", ha='center', fontsize=10)

# Elrendezés javítása
plt.subplots_adjust(bottom=0.2)
plt.show()
```

Az ábra a lineáris regressziós modell alapján előre jelzi a csapatok várható pontszámát a 2026-os diáknapokon. Az előrejelzés azt mutatja meg, mely csapatok teljesítménye eltérő lesz a jövő évben, figyelembe véve az elmúlt évek pontszámait. A legmagasabb pontszámot váró csapatok nem feltétlenül azok, amelyek a legtöbbször vettek részt vagy a legtöbb összpontot gyűjtötték össze korábban. Ez jól szemlélteti a teljesítmény dinamikus változását és a potenciális fejlődést a csapatok között.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Skálázás
features = ['predikcio_2026', 'reszvetelek_szama', 'osszpont']
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_pred[features])

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(scaled_data)

df_pred['cluster'] = clusters

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_pred,
    x='predikcio_2026',
    y='osszpont',
    hue='cluster',
    palette='Set2',
    s=100
)

# Címkék hozzáadása a pontokhoz
for i in range(len(df_pred)):
    plt.text(
        df_pred['predikcio_2026'].iloc[i] + 0.2,
        df_pred['osszpont'].iloc[i],
        df_pred['csapat'].iloc[i],
        fontsize=9
    )

plt.title("Klaszterezés csapatok szerint (2026 predikció vs összpont)")
plt.xlabel("Predikció 2026")
plt.ylabel("Összpont az eddigi évek alapján")
plt.legend(title="Klaszter")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Az ábra a K-means klaszterezési módszerrel csoportosítja a csapatokat három klaszterre, figyelembe véve a 2026-ra előre jelzett pontszámokat és az eddig összegyűjtött pontjaikat. A klaszterezés segítségével könnyen felismerhetők azok a csapatok, amelyek hasonló teljesítménymintázatot mutatnak a jövőbeli várakozások és a múltbeli eredmények alapján egyaránt.
